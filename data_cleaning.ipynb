```python
import pandas as pd
import numpy as np
df1=pd.read_csv('/content/amazon_fires.csv',encoding='ISO-8859-1')
```

```python
  df1
```

```python
# DATACLEANING
# 1 RENAMEING THE COLUMN
d={'ano':'year','mes':'month','estado':'state','numero':'number','encontro':'date'}
df1.rename(columns=d,inplace=True)
df1
```

```python
# 2 REMOVEING DUPLICATE VALUES
df1.drop_duplicates(inplace=True)
df1.info()
df1['month'].unique()
```

```python
# 3 changing the data present in a particular column
d={'Janeiro':'january', 'Fevereiro':'febuary', 'MarÃ§o':'march', 'Abril':'april', 'Maio':'may', 'Junho':'june', 'Julho':'july',
'Agosto':'augest', 'Setembro':'september', 'Outubro':'october', 'Novembro':'november', 'Dezembro':'december'}
df1['month']=df1['month'].map(d)
df1
df1['month'].unique()
```

```python
#4 removing unwanted data from particular col
#df['col_name']=df1['col_name'].str.strip('unwanted')

df1['number']=df1['number'].str.strip(' Fires')
df1
df1.info()
```

```python
df1
```

```python
#chaning the object to float value
df1['number']=df1['number'].astype(float)
df1.info()
```
